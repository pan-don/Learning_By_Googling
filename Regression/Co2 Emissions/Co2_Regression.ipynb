### 1. Supervised Learning

supervised learning adalah metode machine learning yang paling umum. Dalam bahasa indonesia, supervised learning diartikan sebagai pembelajaran diawasi. Istilah "diawasi" ini muncul karena algoritma ini dirancang untuk belajar melalui contoh. Dalam metode ini, model dilatih menggunakan dataset yang sudah berlabel. Ini berarti setiap data input memiliki output yang diketahui, sehingga model dapat belajar dari contoh-contoh tersebut. Supervised Learning dibagi menjadi 2 macam yaitu `regresi` dan klasifikasi

### 2. Unsupervised Learning

Unsupervised Learning merupakan metode mechine learning yang tidak diawasi dimana model dilatih menggunakan data yang tidak memiliki label atau hasil yang diketahui. Dalam metode ini, Model harus menemukan pola atau karakteristik dalam data tanpa bantuan jawaban. Unsupervised Learning biasanya digunakan dalam tugas clustering (mengelompokkan data yang mirip) dan asosiasi (menemukan hubungan antar fitur dalam dataset).

### 3. Reinforcement Learning

Reinforcement Learning aadalah metode mechine learning dimana model belajar melalui interaksi dengan lingkungannya. Dalam metode ini, model belajar melalui trial-and-error, yaitu mencoba berbagai aksi dan mendapatkan "reward" (penghargaan) atau "punishment" (hukuman) sesuai dengan tindakan yang diambil. Tujuan utama model dalam Reinforcement Learning adalah memaksimalkan total reward melalui proses pembelajaran.

regresi adalah metode analisis untuk memodelkan hubungan antara variabel dependen (variabel yang kita prediksi atau hasil) dan satu atau lebih variabel independen (faktor-faktor yang kita gunakan untuk memprediksi hasil). Regresi memungkinkan kita untuk menganalisis kekuatan hubungan antara variabel dependen dan independen. selain itu, dengan menggunakan regresi kita dapat menentukan seberapa besar perubahan dalam variabel independen dapat mempengaruhi variabel dependen. Salah satu jenis regresi yang paling sering digunakan adalah regresi linear.

Regresi termasuk kedalam Statistika Inferensial: Berurusan dengan pengambilan kesimpulan, prediksi, atau keputusan tentang populasi berdasarkan sampel data

Tujuan utama dari regresi adalah menemukan hubungan atau pola antara variabel input (independen) dan output (dependen), sehingga kita dapat memperkirakan nilai output pada data baru.

Contoh regresi dalam sehari-hari:
>- Pengaruh penggunaan TV terhadap penggunaan listrik
>- pengaruh usia mobil terhadap harga jual mobil bekas
>- pengaruh luas tanah terhadap harga rumah
>- pengaruh berat badan terhadap tekanan darah

### Regresi Positif

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
luas_tanah = [60, 80, 100, 120, 140, 160, 180, 200, 220, 240]
harga_rumah = [100000, 150000, 200000, 250000, 300000, 350000, 400000, 450000, 500000, 550000]

# Menghitung koefisien regresi linear (slope dan intercept)
m, b = np.polyfit(luas_tanah, harga_rumah, 1)

# Membuat scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(luas_tanah, harga_rumah, color='green')
plt.plot(luas_tanah, [m * x + b for x in luas_tanah], color='red', linestyle='--')
plt.annotate("Semakin luas tanah, semakin tinggi\n pula harga jual rumah", 
             xy=(150, 330000), xytext=(70, 450000), 
             fontsize=10, color="darkred", 
             bbox=dict(facecolor='white', alpha=0.7),
             arrowprops=dict(facecolor='black', arrowstyle="->"))

# Label dan judul plot
plt.title('Regresi Positif')
plt.xlabel('Luas Tanah (m)')
plt.ylabel('Harga Rumah ($)')
plt.grid(True)
plt.show()
```

## Outliers Handling

Melihat data outlier(data yang jauh dari pola umum kebanyakan data) untuk ditangani lebih lanjut.

```python
# Menangani data outlier dengan mengubah nilai outlier ke nilai kuartil terdekat(Kuartil 1 dan Kuartil 3).

def iqr_outliers(data, col):
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    upper_bound = Q3 + 1.5 * IQR
    lower_bound = Q1 - 1.5 * IQR
    data[col] = data[col].mask(data[col] < lower_bound, Q1)
    data[col] = data[col].mask(data[col] > upper_bound, Q3)
    return data

for col in data.columns:
    iqr_outliers(data, col)
```

## Linear Regression

**Linear Regression** adalah metode statistik yang digunakan untuk menganalisis hubungan antara satu variabel independen (fitur) dengan variabel dependen (target) dengan cara menemukan garis terbaik yang meminimalkan selisih antara prediksi dan nilai aktual. Model ini berasumsi bahwa hubungan antara variabel adalah linear, yang berarti dapat digambarkan dengan garis lurus.
$$\ y = mx + b$$
   > - **y** adalah variabel dependen (target) yang ingin diprediksi.
   > - **x** adalah variabel independen (fitur).
   > - **m** adalah **kemiringan garis** (slope), yang menunjukkan seberapa besar perubahan pada y untuk setiap unit perubahan pada x.
   > - **b** adalah **intercept** (titik potong pada sumbu y), yaitu nilai y ketika x = 0. Ini menggambarkan titik di mana garis regresi berpotongan dengan sumbu y.

```python
# import model
from sklearn.linear_model import LinearRegression

# inisiasi model linear regression
model_linear = LinearRegression()

# melatih model dengan data training
model_linear.fit(x_train, y_train)

# menguji model dengan memprediksi data test
y_pred_lr = model_linear.predict(x_test)
```

```python
# nilai intercept(b) pada model
model_linear.intercept_
```

```python
# nilai slope(m) pada model
linear_coef = pd.DataFrame({'Columns':x.columns, 'Coefficient':model_linear.coef_})
linear_coef
```

```python
# visualisasi scatterplot untuk mengevaluasi kinerja model regresi dalam memprediksi data

plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred_lr, color='green', label='Test data', alpha=0.5, s=10)
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', label='Garis linear', linestyle= '--')
plt.title('Linear Regression Model')
plt.xlabel('Actual Data')
plt.ylabel('Predicted Data')
plt.legend()
plt.grid(True)
plt.show()
```

## Evaluasi Model

Evaluasi model adalah proses mengukur seberapa baik model melakukan prediksi atau pengklasifikasian berdasarkan data baru yang belum pernah dilihatnya (data uji). Dengan mengevaluasi kinerja model, kita dapat mengetahui apakah model yang kita buat sudah sesuai dan layak digunakan atau tidak.

Ada beberapa metriks atau metode untuk mengevalausi model regresi seperti:
> - **Mean Absolute Error** (MAE): mengukur rata-rata dari selisih absolut antara prediksi model dan nilai yang diamati:
$$\frac 1n\sum_{i=1}^n|y_i-\hat{y}_i|$$
    - Semakin rendah nilai MAE, semakin baik model dalam memprediksi nilai target

> - **Mean Squared Error** (MSE) mengukur rata-rata dari kuadrat selisih antara prediksi model dan nilai yang diamati:
$$\frac 1n\sum_{i=1}^n(y_i-\hat{y}_i)^2$$
    - MSE memberikan bobot lebih besar pada perbedaan yang besar antara prediksi dan nilai sebenarnya.
    - Nilai MSE yang lebih rendah menunjukkan kinerja model yang lebih baik

> - **Root Mean Squared Error** (RMSE) dalah akar kuadrat dari MSE dan memberikan interpretasi yang lebih intuitif karena dalam satuan yang sama dengan variabel target:
$$\sqrt{\frac 1n\sum_{i=1}^n(y_i-\hat{y}_i)^2}$$
    - RMSE juga memberikan bobot lebih besar pada perbedaan yang besar antara prediksi dan nilai sebenarnya.
    - Seperti MSE, semakin rendah nilai RMSE, semakin baik kinerja model

> - **R-squared (R) Score** mengukur seberapa baik model regresi mampu menjelaskan variabilitas data, menunjukkan seberapa baik data yang diamati cocok dengan prediksi model:
$$R^2 = 1 - \frac{\sum_{i=1}^n (y_i - \hat{y}_i)^2}{\sum_{i=1}^n (y_i - \bar{y})^2}$$
    - R berkisar antara 0 hingga 1; nilai lebih tinggi menunjukkan bahwa model mampu menjelaskan variansi data dengan lebih baik.
    - Nilai **1** berarti model sangat baik dalam menjelaskan variabilitas data (fit sempurna), sedangkan nilai **0** berarti model gagal menjelaskan variabilitas, sehingga hanya menghasilkan rata-rata data tanpa pola.

```python
# import library yang digunakan untuk evaluasi model
from sklearn.metrics import mean_absolute_error, mean_squared_error, root_mean_squared_error, r2_score

mae_linear = mean_absolute_error(y_test, y_pred_lr)
mse_linear = mean_squared_error(y_test, y_pred_lr)
rms_linear = root_mean_squared_error(y_test, y_pred_lr)
r2_linear = r2_score(y_test, y_pred_lr)

print('mean absolute error:', mae_linear)
print('mean squared error:', mse_linear)
print('root mean squared error:', rms_linear)
print('r2 score:', r2_linear)
```

```python
# visualisasi distribusi error antara nilai aktual dan prediksi model
error_values_lr = y_test - y_pred_lr

plt.figure(figsize=(8, 6))
plt.hist(error_values_lr, bins=50, density=True, color='blue', alpha=0.5)
plt.title('Error Values of Linear regression')
plt.xlabel('Error Values')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```
